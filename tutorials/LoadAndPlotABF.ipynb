# Loading and plotting axon binary files (.abf)

Uses pyabf to load an .abf file and matplotlib to graph the data.

Check out the [pyabf documentation](https://swharden.com/pyabf/tutorial/) to learn more!

```python
import pyabf
import os
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
```

## Load the abf file

```python
fpath = os.path.join(*['..','test_data', 'invitro', '21520000.abf'])
data = pyabf.ABF(fpath)
```

## Get summary of file contents

```python
print(data.headerText)
```

## Loading by sweeps and channels

```python
data.setSweep(sweepNumber=13, channel=0)
t_c0 = data.sweepX
v_c0 = data.sweepY

data.setSweep(sweepNumber=13, channel=2)
t_c1 = data.sweepX
v_c1 = data.sweepY
```

## Plot the data

```python
fig, ax =plt.subplots(figsize=(8,3))
ax.plot(t_c0, v_c0, label='Ch0')
ax.plot(t_c1, v_c1, label='Ch1')
ax.set_ylabel('V (mV)')
ax.set_xlabel('Time (s)')
ax.legend()
plt.show()
```
